```python
import math
import numpy as np
import tkinter as tk
import matplotlib.pyplot as plt
```

```python
line_gap = int(input())
```

```python
r1x = []
r1y = []
r2x = []
r2y = []


for i in range(0,20):
  r1x.append(i)
  r1y.append(0)
  r2x.append(i)
  r2y.append(line_gap)
  
```

```python
print("How many needle? And line length?")
needle_number = int(input())
length = int(input())
```

```python
xmin, xmax = [0, 20]
ymin, ymax = [0, line_gap]

hit = 0

plt.figure(figsize=(10,2))
plt.plot(r1x,r1y)
plt.plot(r2x,r2y)

for i in range(0, needle_number):
  x1 = np.random.uniform(xmin, xmax)
  y1 = np.random.uniform(ymin, ymax)
  angle = np.random.uniform(0, 2*np.pi)
  x2 = x1 + length * np.cos(angle)
  y2 = y1 + length * np.sin(angle)
  plt.plot([x1, x2], [y1, y2])

  if (y1 < 0 < y2) or (y1 > 0 > y2):
    hit = hit +1 
  if (y1 < line_gap < y2) or (y1 > line_gap > y2):
    hit = hit + 1


plt.show()
print(hit, " hits")
pi = (2 * length * needle_number) / ( line_gap * hit)
print("Pi = ",pi)
```
